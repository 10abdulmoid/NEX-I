```python
import cv2
from ultralytics import YOLO
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders
import matplotlib.pyplot as plt

person_model = YOLO('yolov8n.pt')  # Person detection model
weapon_model = YOLO('trained_gun_model.pt')  # Trained weapon detection model

# Email setup function
def send_alert(frame):
    sender_email = "saadriyanedu@gmail.com"
    receiver_email = "mohammedsaadriyan@gmail.com"
    subject = "Security Alert: Person with Weapon Detected"
    body = "A person with a weapon was detected. Please check the attached image."

    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = receiver_email
    msg['Subject'] = subject
    msg.attach(MIMEText(body, 'plain'))

    # Save frame as image and attach
    img_path = '/content/alert.jpg'
    cv2.imwrite(img_path, frame)

    attachment = open(img_path, 'rb')
    part = MIMEBase('application', 'octet-stream')
    part.set_payload(attachment.read())
    encoders.encode_base64(part)
    part.add_header('Content-Disposition', f"attachment; filename= alert.jpg")
    msg.attach(part)

    # Email server configuration
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login(sender_email, 'your_password')  # Replace with your email password
    text = msg.as_string()
    server.sendmail(sender_email, receiver_email, text)
    server.quit()

# Path to the uploaded video file
video_path = '/content/robbery.mp4'  # Replace with the actual filename

# Attempt to open the video file
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Error: Could not open video file {video_path}.")
    exit()

# Get the width, height, and fps of the frames
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Define the codec and create a VideoWriter object
output_path = '/content/output_video.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Process the video frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLOv8 models on the frame
    person_results = person_model(frame)
    weapon_results = weapon_model(frame)

    person_detected = False
    weapon_detected = False

    # Process person detection results
    for result in person_results:
        boxes = result.boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0]
            score = box.conf[0]
            label = box.cls[0]
            if label == 0:  # Label 0 is for 'person' in COCO dataset
                person_detected = True
                # Draw bounding box
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (255, 0, 0), 2)
                # Put label
                cv2.putText(frame, f'Person {score:.2f}', (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # Process weapon detection results
    for result in weapon_results:
        boxes = result.boxes
        for box in boxes:
            x1, y1, x2, y2 = box.xyxy[0]
            score = box.conf[0]
            label = box.cls[0]
            if label == 1:  # Assuming label 1 is for 'weapon' in your trained model
                weapon_detected = True
                # Draw bounding box
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 0, 255), 2)
                # Put label
                cv2.putText(frame, f'Weapon {score:.2f}', (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)

    # Send alerts if both person and weapon are detected in the same frame
    if person_detected and weapon_detected:
        send_alert(frame)

    # Write the frame into the file
    out.write(frame)

# Release the video capture and writer objects
cap.release()
out.release()

print("Processing complete. The video is saved as output_video.mp4.")
```
